### 1.1 IMDB Scraping

```python
def IMDB_scrape_movie(movie_code):
    rating_list = []
    index = 0
    
    while True:
        page=requests.get("http://www.imdb.com/title/"+movie_code+"/reviews-index?start="+str(index))
        soup = BeautifulSoup(page.text, "html.parser")
        reviews = soup.find_all("td", attrs= {"class":"comment-summary"})
        
        #if there are no more reviews to be scraped
        if len(reviews) == 0:
            break
        counter = 0
        
        #get the scores for each review
        for review in reviews:
            if review.find("img", attrs={"width":102}):
                rating_list.append(int(review.find("img", attrs={"width":102})['alt'].split('/')[0]))
        index = index+50
    return rating_list
```

### 1.2 Metacritic Scraping

```python
def metacritic_scrape_movie(movie_code):
    rating_list = []
    index = 0
    
    while True:    
        page=requests.get("http://www.metacritic.com/movie/"+movie_code+"/user-reviews?sort-by=date&num_items=100&page=" + str(index), headers={'User-Agent': 'Mozilla/5.0'})
        soup = BeautifulSoup(page.text, "html.parser")
        
        #if there are no more reviews, break
        if soup.find("div", attrs= {"class":"msg msg_no_reviews"}):
            break
        
        
        table = soup.find("ol", attrs= {"class":"reviews user_reviews"})  
        review_list = table.find_all("div", attrs= {"class":"review_grade"})
        for review in review_list:
            rating_list.append(int(review.find("div").get_text()))
        index+=1
    
    return rating_list
```

### 1.3 Fandango

```python
def fandango_scrape_movie(movie_code):
    rating_list = []
    index = 1
   
    while True: 
        page=requests.get("http://www.fandango.com/"+movie_code+"/moviereviews?pn=" + str(index))
        soup = BeautifulSoup(page.text, "html.parser")

        #check if there are any more reviews
        if soup.find("div", attrs={"class":"no-fan-review-message-container"}):
            break

        reviews = soup.find_all("div", attrs={"class":"movie-review-item critic-review-content decoration-bottom"})
        for review in reviews:
            num_stars = len(review.find_all("span", attrs={"class":"star-icon full"}))
            rating_list.append(num_stars)
        index+=1
    return rating_list
```

### 1.4 Rotten Tomatoes

```python
def rotten_tomatoes_scrape_movie(movie_code):
    rating_list = []
    index = 1
    while True:
        page=requests.get("http://www.rottentomatoes.com/m/"+movie_code+"/reviews/?page="+str(index)+"&type=user&sort=")
        soup = BeautifulSoup(page.text, "html.parser")

        if not soup.find("div", attrs={"class":"review_table"}):
            break

        page_reviews = soup.find("div", attrs={"class":"review_table"}).find_all("div", attrs={"class":"row review_table_row"})


        for review in page_reviews:
            review_stars = review.find("span", attrs={"class":"fl"})
            
            if not review_stars:
                continue
            #Check get the number of stars
            stars = review_stars.find_all("span", attrs ={"class":"glyphicon glyphicon-star"})

            num_stars = len(stars)
            if len(review_stars.get_text()) > 1:
                num_stars += 0.5
            rating_list.append(num_stars)
        index +=1

    return rating_list
```

## 3. Partial Pooling using Rotten Tomatoes Data

In this sub-problem, we use the data from only Rotten Tomatoes. Hence, it is only a 2-level hierarchical Bayes model. The movie reviews from Rotten Tomatoes for the 10 movies are used as an observation of how good the studio is. We model each studio as having a normal distribution with a mean $\alpha_{studio}$ and variance $\beta_{studio}$. $\alpha_{studio}$ and $\beta_{studio}$ have uniform priors. Each movie is modeled as a normal distribution where the mean $\theta_j$ drawn from $N(\alpha_{studio},\beta_{studio})$.  We assume that the standard deviation of each movie in the model is due to the different sample sizes.

### 3.1 Dreamworks

```python
# Here is the data
rt_list = []
for movie in dreamworks_raw['fandango']:
    rt_list.append((np.mean(dreamworks_raw['fandango'][movie]), np.std(dreamworks_raw['fandango'][movie])))
    
data = np.double(np.array(rt_list))
data_means = data[:,0]
data_std = data[:,1]

theta_list = []


alpha = pm.Uniform("alpha", -0., 10.0)
beta = pm.Uniform("beta", 2., 20.0)

for x in range(len(data_means)):
    theta = pm.Normal("theta"+str(x), mu = alpha, tau = 1./beta)

    theta_list.append(theta)

    
print len(theta_list)

#Normal
likelihood = pm.Normal('likelihood',mu=[theta_list], tau=1./data_std**2,value=data_means,observed=True)

parameterlist = [alpha, beta, likelihood]
parameterlist.extend(theta_list)
response_model=pm.Model(parameterlist)
mcmc = pm.MCMC(response_model)
mcmc.sample(iter=100000, burn=10000, thin = 2)
```

```python
a_hist = plt.hist(alpha.trace(), bins=50)
a_map_index = np.argmax(a_hist[0])
a_hist_bin_middle = 0.5*a_hist[1][:-1] + 0.5*a_hist[1][1:]
a_map = a_hist_bin_middle[a_map_index]
print "Alpha MAP value is:",a_map
```

## 4. Partial Pooling with 4 websites

Now, we extend the Hierarchical Bayes model to another level. Rather than using data from only Rotten Tomatoes, we use data from 4 different websites (Rotten Tomatoes, IMDB, Metacritic and Fandango). Each of the 4 website reviews for a particular movie is an observation of how good that movie is. Recall from the section above that each movie is modeled as a normal distribution where the mean $\theta_j$ drawn from $N(\alpha_{studio},\beta_{studio})$. Now, we give each movie a uniform prior for its variance, $\rho_j$. Therefore, each website review is modeled as a normal distribution where the mean $\mu$ is drawn from $N(\theta_j, \rho_j)$ 

### 4.1 Dreamworks

```python
dreamworks = {}

for website in dreamworks_raw:
    temp_list = []
    for movie in dreamworks_raw[website]:
        temp_list.append((np.mean(dreamworks_raw[website][movie]), np.std(dreamworks_raw[website][movie])))
    dreamworks[website] = list(temp_list)

pixar = {}

for website in pixar_raw:
    temp_list = []
    for movie in pixar_raw[website]:
        temp_list.append((np.mean(pixar_raw[website][movie]), np.std(pixar_raw[website][movie])))
    pixar[website] = list(temp_list)

```

```python
dreamworks_list = []


for x in range(10):
    for movie in dreamworks:
        dreamworks_list.append([dreamworks[movie][x][0], dreamworks[movie][x][1]])

dreamworks_list = np.array(dreamworks_list)
dreamworks_mean =  dreamworks_list[:,0]
dreamworks_std =  dreamworks_list[:,1]

```

```python
NUM_MOVIES = 10
NUM_WEBSITE = 4


alpha = pm.Uniform("alpha", 0., 10.0)
beta = pm.Uniform("beta", 2., 20.0)
theta_list = []
for x in range(NUM_MOVIES):
    theta = pm.Normal("theta"+str(x), mu = alpha, tau = 1./beta)

    theta_list.append(theta)

std_list = []
for x in range(NUM_MOVIES):
    movie_std = pm.Uniform("website_std"+str(x), 2., 20.0)

    std_list.append(movie_std)
    
#add another level of the hierarchial bayes
websites_list = []
for x in range (NUM_MOVIES):
    for y in range (NUM_WEBSITE):
        websites = pm.Normal("website"+str(x*NUM_WEBSITE+y), mu = theta_list[x], tau =1./std_list[x])
        websites_list.append(websites)

@pm.observed
def likelihood(mu=websites_list, tau=1./dreamworks_std**2,value=dreamworks_mean):
    likelihood = 0
    for i in range(len(data_means)):
        likelihood += pm.normal_like(value[i],mu[i],tau[i])
    return likelihood

parameterlist = [alpha, beta, likelihood]
parameterlist.extend(theta_list)
response_model=pm.Model(parameterlist)
mcmc = pm.MCMC(response_model)
mcmc.sample(iter=100000, burn=10000, thin = 2)
```

```python
a_hist = plt.hist(alpha.trace(), bins=50)
a_map_index = np.argmax(a_hist[0])
a_hist_bin_middle = 0.5*a_hist[1][:-1] + 0.5*a_hist[1][1:]
a_map = a_hist_bin_middle[a_map_index]
print "Alpha MAP value is:",a_map
print "Quantiles:",mcmc.stats()['alpha']['quantiles']
```

### 4.2 Pixar

```python
NUM_MOVIES = 10
NUM_WEBSITE = 4


alpha2 = pm.Uniform("alpha2", 0., 10.0)
beta2 = pm.Uniform("beta2", 2., 20.0)
theta_list = []
for x in range(NUM_MOVIES):
    theta = pm.Normal("theta"+str(x), mu = alpha2, tau = 1./beta2)

    theta_list.append(theta)

std_list = []
for x in range(NUM_MOVIES):
    movie_std = pm.Uniform("website_std"+str(x), 2., 20.0)

    std_list.append(movie_std)
#add another level of the hierarchial bayes
websites_list = []
for x in range (NUM_MOVIES):
    for y in range (NUM_WEBSITE):
        websites = pm.Normal("website"+str(x*NUM_WEBSITE+y), mu = theta_list[x], tau =1./std_list[x])
        websites_list.append(websites)

@pm.observed
def likelihood(mu=websites_list, tau=1./pixar_std**2,value=pixar_mean):
    likelihood = 0
    for i in range(len(data_means)):
        likelihood += pm.normal_like(value[i],mu[i],tau[i])
    return likelihood

parameterlist = [alpha2, beta2, likelihood]
parameterlist.extend(theta_list)
response_model=pm.Model(parameterlist)
mcmc = pm.MCMC(response_model)
mcmc.sample(iter=100000, burn=10000, thin = 2)
```

```python
a_hist = plt.hist(alpha2.trace(), bins=50)
a_map_index = np.argmax(a_hist[0])
a_hist_bin_middle = 0.5*a_hist[1][:-1] + 0.5*a_hist[1][1:]
a_map = a_hist_bin_middle[a_map_index]
print "Alpha MAP value is:",a_map
print "Quantiles:",mcmc.stats()['alpha2']['quantiles']
```
